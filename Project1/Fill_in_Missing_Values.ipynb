##Filling FULLVAL Null Values

```
nafullval=df[df['FULLVAL'].isna()]
nafullval
```

```
countzip1=df.groupby(['ZIP','TAXCLASS','BLDGCL']).count()
countzip1=countzip1.reset_index()
countzip1=countzip1[countzip1['RECORD']>5]
countzip1=countzip1[['ZIP', 'TAXCLASS','BLDGCL']].copy()
countzip1 = countzip1[countzip1.ZIP != 'nan']
countzip1
```

```
ziplv1=df.groupby(['ZIP','TAXCLASS','BLDGCL']).median()
ziplv1=ziplv1.dropna().reset_index()
ziplv1
```

```
fvlv1=ziplv1[['TAXCLASS', 'ZIP','BLDGCL','FULLVAL']].copy()
fvlv1.rename(columns = {'FULLVAL':'Lv1FULLVAL'}, inplace = True) 
fvlv1
```

```
fvlv1 = pd.merge(countzip1, fvlv1,  how='inner', left_on=['ZIP', 'TAXCLASS','BLDGCL'], right_on = ['ZIP', 'TAXCLASS','BLDGCL'])
fvlv1=fvlv1[['TAXCLASS', 'ZIP','BLDGCL','Lv1FULLVAL']].copy()
fvlv1
```

```
df = pd.merge(df, fvlv1,  how='left', left_on=['TAXCLASS','ZIP','BLDGCL'], right_on = ['TAXCLASS','ZIP','BLDGCL'])
```

```
df.FULLVAL[df['FULLVAL'].isna()] = df.Lv1FULLVAL
```

```
df.drop(['Lv1FULLVAL'], axis=1,inplace=True)
df[df['FULLVAL'].isna()]
```

```
level1result=len(nafullval)-len(df[df['FULLVAL'].isna()])
```

```
print('There are ',level1result,' null values filled.')
```

There are  153  null values filled with Zip5 data, while with Zip3 there are 487 values filled. Now the missing values are shown as Na.

Level 2 Filling

```
countzip2=df.groupby(['TAXCLASS','ZIP']).count()
countzip2=countzip2.reset_index()
countzip2=countzip2[countzip2['RECORD']>5]
countzip2=countzip2[['TAXCLASS','ZIP']].copy()
countzip2
```

```
lvzip2=df.groupby(['TAXCLASS','ZIP']).median()
lvzip2=lvzip2.reset_index()
lvzip2=lvzip2[['TAXCLASS','ZIP','FULLVAL']].copy()
lvzip2
```

```
lvzip2 = pd.merge(lvzip2, countzip2,  how='inner', left_on=['TAXCLASS','ZIP'], right_on = ['TAXCLASS','ZIP'])
lvzip2
lvzip2.rename(columns = {'FULLVAL':'L2FV'}, inplace = True) 
lvzip2
```

```
df = pd.merge(df, lvzip2,  how='left', left_on=['TAXCLASS','ZIP'], right_on = ['TAXCLASS','ZIP'])
```

```
df.FULLVAL.fillna(df.L2FV, inplace=True)
```

```
level2result=len(nafullval)-len(df[df['FULLVAL'].isna()])-level1result
```

```
print('There are a total of',len(nafullval)-len(df[df['FULLVAL'].isna()]),' null values filled, and this level filled ',level2result,' values.')
```

Level 3 Filling:
* This level of filling will be updating median value to all null values with the overall median of the full value.

```
df['FULLVAL']=df['FULLVAL'].fillna(df.groupby(['ZIP'])['FULLVAL'].transform('median'))
df[df['FULLVAL'].isna()]
```
