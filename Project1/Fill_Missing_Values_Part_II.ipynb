```python
new_df['LOTAREA'] = new_df['LTFRONT']*new_df['LTDEPTH']
new_df
```

```python
new_df['BLDAREA'] = new_df['BLDFRONT']*new_df['BLDDEPTH']
new_df['BLDVOL'] = new_df['BLDAREA']*new_df['STORIES']
new_df.head(3)
```

```python
df45 = pd.DataFrame(columns = ['r1','r2','r3','r4','r5','r6','r7','r8','r9'])
df45
```

```python
df45['r1'] = new_df['FULLVAL']/new_df['LOTAREA']
```

```python
df45['r2'] = new_df['FULLVAL']/new_df['BLDAREA']
df45['r3'] = new_df['FULLVAL']/new_df['BLDVOL']
df45['r4'] = new_df['AVLAND']/new_df['LOTAREA']
df45['r5'] = new_df['AVLAND']/new_df['BLDAREA']
df45['r6'] = new_df['AVLAND']/new_df['BLDVOL']
df45['r7'] = new_df['AVTOT']/new_df['LOTAREA']
df45['r8'] = new_df['AVTOT']/new_df['BLDAREA']
df45['r9'] = new_df['AVTOT']/new_df['BLDVOL']
```

```python
df45.head(5)
```

```python
groups = ['ZIP5', 'ZIP3', 'B', 'TAXCLASS']
rs = ['r1','r2','r3','r4','r5','r6','r7','r8','r9']
for r in rs:
    for g in groups:
        df45[r+g] = df45[r]/df45.groupby([g])[r].transform('mean')
```
